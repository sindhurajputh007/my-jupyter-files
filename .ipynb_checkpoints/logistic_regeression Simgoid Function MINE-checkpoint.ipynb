```python
print(f"Enter 1 for positive class or 0 for negative class.")
y = int(input())

if y==1:
    print(f"You have select positive case.")
else:
    print(f"You have selected negative case. ")


print(f"Enter the estimated probability between 0 and 1.")
p = float(input())

print(f"The estimated probability you have provided is {p}.")

if y == 1:
    c = -np.log(p)
    print(f'Cost of predicting {p} is {c} when y is positive or y=1.')
else:
    c = -np.log(1- p)
    print(f'Cost of predicting {p} is {c} when y is negative or y=0.')
```

```python
from sklearn.datasets import load_iris
from sklearn import tree

iris = load_iris()
X, y = iris.data, iris.target
clf = tree.DecisionTreeClassifier()
clf = clf.fit(X, y)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(40,20)) 
tree.plot_tree(clf)
plt.show()
```
