```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import seaborn as sns

#load the iris dataset
iris = sns.load_dataset("iris")

#create a 3D scatter plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(iris['sepal_length'], iris['sepal_width'], iris['petal_length'])
ax.set_xlabel('Sepal Length')
ax.set_ylabel('Sepal Width')
ax.set_zlabel('Petal Length')
plt.show()
```

# PRBABILITY FUNCTION

```python
#compute the mean and standard deviation of sepal length for setosa flowers
setosa_sepal_length_mean = np.mean(iris.loc[iris['species']== 'setosa', 'sepal_length'])
setosa_sepal_length_std = np.std(iris.loc[iris['species'] == 'setosa', 'sepal_length'])

#Define the gaussian probability function
def gaussian(x, mu, sigma):
    return (1/ (sigma * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - mu) / sigma) **2)

#compute the probability of an iris flower with sepla length 5 belonging to the setosa species
x = 5
p_setosa_gives_sepal_length_5 = gaussian(x, setosa_sepal_length_mean, setosa_sepal_length_std)

print(p_setosa_given_sepal_length_5)
```

# UNIVARIATE ANALYSIS

```python
#compute the mean and standard deviation of sepal lengths
sepal_length_mean = np.mean(iris['sepal_length'])
sepal_length_std = np.std(iris['sepal_length'])

#define the gaussian probability function
def gaussian(x, mu, sigma):
    return(1 / (sigma * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - mu) / sigma) ** 2)

#compute the probability density function for sepal lengths using gaussian distribution
x_values = np.linspace(3, 9, 100)
pdf_values = gaussian(x_values, sepal_length_mean, sepal_length_std)

#plot the probability density function
sns.lineplot(x=x_values, y=pdf_values)
```

# CUMULATIVE DISTRIBUTION FUNCTION

```python
#compute the CDF for sepal length
sepal_length_cdf = np.cumsum(np.histogram(iris['sepal_length'], bins=10, density=True)[0])

#plot the CDF
sns.lineplot(x=np.histogram(iris['sepal_length'], bins=10)[1][1:], y=sepal_length_cdf)
```

## VARIANCE & STANDARD DEVIATION

```python
# compute the variance and standard deviation of sepal length
sepal_length_var = np.var(iris['sepal_length'])
sepal_length_std = np.std(iris['sepal_length'])

#print the results
print("sepal length variance:", sepal_length_var)
print("sepal length standard deviation:", sepal_length_std)
```

## MEDIAN, PERCENTILES & QUARTILES


```python
# compute the median, percentiles, and quartiles od sepal length
sepal_length_median = np.median(iris['sepal_length'])
sepal_length_percentiles = np.percentile(iris['sepal_length'], [25, 50, 75])
sepal_length_quartiles = np.quantile(iris['sepal_length'], [0.25, 0.5, 0.75])

#print the results
print("sepal length median:", sepal_length_median)
print("sepal_length percentiles:", sepal_length_percentiles)
print("sepal_length quartiles:", sepal_length_quartiles)
```
